```python
import numpy as np
import pandas as pd
import wooldridge as wd
```

```python
wage = wd.data('wage1')[['wage','educ','exper']]
wage
```

```python
y = wage['wage'].to_numpy()
y
```

```python
x = wage[['educ','exper']].to_numpy()
x
```

```python
x.T
```

```python
x.T.dot(x)
```

```python
np.linalg.inv(x.T.dot(x))
```

```python
betas=np.linalg.inv(x.T.dot(x)).dot(x.T.dot(y))
betas
```

```python
x.dot(betas)
```

```python
y_est = x.dot(betas)
```

```python
y- y_est
```

```python
u_est = y - y_est
u_est
```

```python
# para saber el numero de datos 
y.shape
```

```python
n,k=x.shape
n,k
```

```python
sum(u_est**2)
```

```python
desv_est=sum(u_est**2)/(n-k)
desv_est
```

```python
x.T
```

```python
x.T.dot(x)
```

```python
ee =np.diag(np.sqrt(np.linalg.inv(x.T.dot(x))*desv_est))
ee
```

```python
betas/ee
```

```python
t = betas/ee
t
```

```python
int_confianza_sup = betas + 1.96*(ee)
int_confianza_sup
```

```python
int_confianza_inf= betas - 1.96*(ee)
int_confianza_inf
```
